```python
#Importing stopwords from nltk.corpus
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
#[^A-Za-z]+ selects any character that matches the rule inside of the square bracket ([]), 
#that does not (^) have at least one (+) letter in upper case alphabets (A-Z) or lower case alphabets (a-z). 
#Then the regex sub replaces these characters in the old text with space.
rev=re.sub('[^a-zA-Z]',' ',Data['messages'][43])
```

```python
rev=rev.lower()
rev
```

```python
from nltk.corpus import stopwords
rev=[ps.stem(word) for word in rev if not word in stopwords.words('english')]
```

```python
#Writing the function
corpus=[]
for i in range(0,len(Data)):
    review=re.sub('[^a-zA-Z]','',Data['messages'][i]) #replace unnecessary symbols with space
    review=review.lower() #Lowering the review column
    review=review.split() #Splitting the data
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
    
```

```python
corpus
```

```python
stopwords.words('english')
```

```python
#Importing the countvectorizer from sklearn
from sklearn.feature_extraction.text import CountVectorizer
```

```python
CV=CountVectorizer(max_features=2500)
```

```python
x=CV.fit_transform(corpus).toarray()
```

```python
x[30]
```

```python
x.shape
```

```python
Data['label']
```

```python
#Importing the y variable by using dummies
y=pd.get_dummies(Data['label'],drop_first=True)
```

```python
x
```

```python
#train test split from sklearn
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33, random_state=42)
```

```python
#Model building
#Here we are creating the model of gaussians NB from sklearn naive bayes
from sklearn.naive_bayes import GaussianNB
```

```python
model1=GaussianNB()
```

```python
model1.fit(x_train,y_train)
```

```python
#Predicting the y_test
y_predict1=model1.predict(x_test)
```

```python
#Importing the accuracy_score from sklearn metrics
from sklearn.metrics import accuracy_score
```

```python
#Predicted value
GB=accuracy_score(y_test,y_predict1)
GB
```
